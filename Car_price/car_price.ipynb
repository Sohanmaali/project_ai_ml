```python
import pandas as pd
import numpy as np
```

```python
data  =  pd.read_csv("car_Data.csv")
```

```python
data.info()
```

```python
data["fuel_type"] = data["fuel_type"].fillna("Petrol")

data.info()
```

```python
print(data["kms_driven"].unique())
```

```python
# Clean the column
data["kms_driven"] = (
    data["kms_driven"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.replace(" kms", "", regex=False)
)

# Convert to numeric
data["kms_driven"] = pd.to_numeric(data["kms_driven"], errors="coerce")

# Fill missing values with median — no inplace warning
data["kms_driven"] = data["kms_driven"].fillna(data["kms_driven"].median())

print(data["kms_driven"].unique())
```

```python
data.info()
```

```python
data["Price"].unique()
```

```python
# Clean the column
data["Price"] = (
    data["Price"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.replace(" kms", "", regex=False)
)

# Convert to numeric
data["Price"] = pd.to_numeric(data["Price"], errors="coerce")

# Fill missing values with median — no inplace warning
data["Price"] = data["Price"].fillna(data["Price"].median())

print(data["Price"].unique())
```

```python
data.info()
```

```python
data.head()
```

```python
data["name"].value_counts()
name_count = data['name'].value_counts()
name_count_less_3 = name_count[name_count <= 1]
name_count_less_3
data["name"] = data["name"].apply(lambda x: "other" if x in name_count_less_3 else x) 
```

```python
data["year"] = (
    data["year"]
    .astype(str)
    .str.strip()
    .str.extract(r"(\d{4})")           # extract 4-digit numbers
)

# Convert to numeric
data["year"] = pd.to_numeric(data["year"], errors="coerce")

# Replace NaN with median year (without inplace)
data["year"] = data["year"].fillna(data["year"].median())

# Convert to int
data["year"] = data["year"].astype(int)

# Check unique sorted years
print(sorted(data["year"].unique()))
```
